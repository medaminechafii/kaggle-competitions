##preprocessing

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train),columns=columns)
X_train.head()
X_test = pd.DataFrame(scaler.transform(X_test),columns=X_test.columns)
```

##Model training

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 200,max_depth = 20)
rf.fit(X_train,y_train)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 200)
lr.fit(X_train,y_train)
```

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators=[("knn",knn),("rf",rf),("xgb",xgb),("lr",lr)],voting = "soft")
vc.fit(X_train,y_train)
```

```python
from sklearn.ensemble import StackingClassifier
sc = StackingClassifier(estimators=[("knn",knn),("rf",rf),("xgb",xgb),("vc",vc),("lr",lr)],final_estimator=KNeighborsClassifier())
sc.fit(X_train,y_train)
```

```python
models = {"knn":knn,"rf":rf,"lr":lr,"xgb":xgb,"voting cl":vc,"stacking cl":sc}
```

```python
def mc_and_acc(models,X_test,y_test,accuracy_score):
  for key in models.keys():
    model = models[key]
    print(f"the acc and mc of the model {key}")
    y_pred = model.predict(X_test)
    print(f"the accuracy is {accuracy_score(y_test,y_pred)}")
```

```python
from sklearn.metrics import roc_auc_score
mc_and_acc(models,X_test,y_test,roc_auc_score)
```
